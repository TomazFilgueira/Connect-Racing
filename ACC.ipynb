```python
df = pd.read_csv('Telemetria Paul Ricard\Paul_Ricard-ferrari_296_gt3-13-2025.02.10-11.43.53.csv',delimiter=',',header=12)
df
```

 Concatenate head + first row in order to achieve variable name + units

```python
df_mod = df.copy()
header = df_mod.columns[:]
first_row = df_mod.iloc[0,:]
df_mod.columns = header + ' (' + first_row + ")" #concatenate header + first row
df_mod  = df_mod.iloc[1:] #drop first row
df_mod
```

Let's investigate columns with zero value

```python
columns = df_mod.columns
#print(columns)

for c in columns:
    print(df_mod[c].value_counts())
    print("\n")
```

```python
# prompt: convert column from object to float and int types

for column in df_mod.columns:
  try:
    df_mod[column] = pd.to_numeric(df_mod[column], errors='coerce')
    if (df_mod[column] % 1 == 0).all():
      df_mod[column] = df_mod[column].astype('Int64')
    else:
      df_mod[column] = df_mod[column].astype('float64')
  except:
    pass

df_mod.info()
```

Plotting Histogram to analyse column distribution


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Determine the number of numeric columns
numeric_columns = [column for column in df_mod.columns if df_mod[column].dtype in ['int64', 'float64']]
num_columns = len(numeric_columns)

# Create subplots
rows = (num_columns + 2) // 3  # Adjust rows for 3 columns per row
fig, axes = plt.subplots(rows, 3, figsize=(15, 5 * rows))  # Adjust figure size
axes = axes.flatten()  # Flatten the axes array for easy iteration

# Plot each histogram
for i, column in enumerate(numeric_columns):
    df_mod[column].plot.hist(ax=axes[i], title=column)
    axes[i].set_xlabel(column)

# Hide unused subplots
for j in range(i + 1, len(axes)):
    axes[j].set_visible(False)

plt.tight_layout()
plt.show()
```

```python
# Select the variables to plot
variables_to_plot = ['G_LAT (m/s2)', 'STEERANGLE (deg)']  # Replace with your column names
num_variables = len(variables_to_plot)

# Create subplots
fig, axes = plt.subplots(num_variables, 1, figsize=(10, 5 * num_variables), sharex=True)

# Plot each variable over time
for i, variable in enumerate(variables_to_plot):
    axes[i].plot(df_mod['Time (s)'], df_mod[variable])  # Replace 'time' with your time column
    axes[i].set_title(f"{variable} over Time")
    axes[i].set_ylabel(variable)

# Set the x-axis label for the last subplot
axes[-1].set_xlabel('Time')

plt.tight_layout()
plt.show()
```
